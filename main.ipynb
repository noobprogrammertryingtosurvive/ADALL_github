### Visualizing Distributions of Numerical Features

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['Price_SGD', 'RAM_GB', 'Storage_GB', 'Weight_kg', 'Screen_Size_inch', 'Discount_percent', 'Brand_Discount', 'Member_Discount']

plt.figure(figsize=(18, 12))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

### Visualizing Distributions of Categorical Features (Top N categories)

```python
categorical_cols = ['Brand', 'Storage_Type', 'Touchscreen']

plt.figure(figsize=(18, 6))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(1, 3, i)
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Count of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
